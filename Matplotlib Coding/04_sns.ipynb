# Box Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
diamonds = pd.read_csv('/Users/puchku-home/Study/Python Libs/Data Viz with Matplotlib & Seaborn/Data/Diamonds Prices2022.csv')

diamonds.head()
```

```python
sns.boxplot(
    data=diamonds,
    x="price"
)
```

```python
sns.boxplot(
    data=diamonds,
    x="price",
    color="orange",
)
```

```python
sns.boxplot(
    data=diamonds,
    y="price",
    color="orange",
)
```

```python
sns.boxplot(
    data=diamonds,
    y="price",
    x="cut",
    palette=["red","blue","purple","orange","green"]
)
```

# Violin Plots

```python
sns.violinplot(
    data=diamonds,
    x="cut",
    y="price",
    palette=["red","blue","purple","orange","green"]
)
```
