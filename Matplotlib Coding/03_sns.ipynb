# Section 1 : Bar Chart and Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.gridspec as gridspec
import numpy as np
import pandas as pd
import matplotlib.patches as mpatches
```

```python
hotels = pd.read_excel('/Users/puchku-home/Study/Python Libs/Data Viz with Matplotlib & Seaborn/Data/HotelCustomersDataset.xlsx')

hotels.head()
```

```python
top5_countries=hotels.query('Nationality in ["DEU", "FRA", "PRT", "GBR", "ESP"]')

top5_countries
```

```python
fig,ax = plt.subplots(1,1)

sns.barplot(
    data=top5_countries,
    x="RoomNights",
    y="Nationality",
    palette="deep",
    ci=None
)
sns.despine()
```

```python
fig,ax = plt.subplots(1,1)

sns.histplot(
    data=hotels.query("LodgingRevenue<1500"),
    x="LodgingRevenue",
    stat="percent"
)

sns.despine()
```

# Assingment 3 : Joint Plot & Pair Plot & Reg Plot & LM Plot

```python
sns.regplot(
    data=top5_countries,
    x="RoomNights",
    y="LodgingRevenue"
)
```

```python
sns.lmplot(
    data=top5_countries.query("RoomNights <=16 and LodgingRevenue<2000"),
    x="RoomNights",
    y="LodgingRevenue",
    hue="Nationality",
)
```

```python
sns.pairplot(
    data=top5_countries.iloc[:, [2, 5, 6, 10, 11, 12]],
    diag_kind="kde",
    corner=True
)
```
