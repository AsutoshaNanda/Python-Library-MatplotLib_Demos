```python
df=df.groupby("region_name").agg({'median_active_list_price': 'sum'})
```

```python
df=df.head()
```

```python
fig ,ax = plt.subplots(figsize=(8,7))

ax.pie(
    x=df["median_active_list_price"],
    labels = df.index,
    startangle=90,
    autopct="%.1f%%"
)
```

```python
fig ,ax = plt.subplots(figsize=(8,7))

ax.pie(
    x=df["median_active_list_price"],
    labels = df.index,
    startangle=90,
    autopct="%.1f%%",
    pctdistance = 0.85
)

donut_hole=plt.Circle((0,0),0.7,facecolor="white")
fig = plt.gcf()
fig.gca().add_artist(donut_hole)
```
