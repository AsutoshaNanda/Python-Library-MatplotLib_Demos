```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = [1,5,11,24,36,48,61,67,75,81]

print(data)
```

```python
labels = [f'{x}*' for x in range (10)]

print(labels)
```

```python
fig , ax = plt.subplots()
ax.plot(labels,data)

fig.suptitle('overall most upper heading title',fontsize =16)
ax.set_title('chart title',fontsize = 12)

ax.set_xlabel('labels random',fontsize =16)
ax.set_ylabel('data used',fontsize =16)

ax.set_xlim()
ax.set_ylim()

ax.axvline(7)
ax.axhline(35)

# .text(x,y,"message")
ax.text(7,35,'point marked',fontsize = 18)

ax.spines[['top','left','right','bottom']]
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(labels,data)
```
