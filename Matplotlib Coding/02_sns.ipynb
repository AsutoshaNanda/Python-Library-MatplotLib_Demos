# Barplots

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
diamonds = pd.read_csv('/Users/puchku-home/Study/Python Libs/Data Viz with Matplotlib & Seaborn/Data/Diamonds Prices2022.csv')

diamonds.head()
```

```python
colors_list=["red","blue","orange","grey","purple"]
sns.barplot(
    data=diamonds,
    x="cut",
    y="carat",
    palette=colors_list,
    ci=None
)
```

```python
sns.barplot(
    data=diamonds,
    x="carat",
    y="cut",
    palette="deep",
    ci=None
)
```

```python
sns.barplot(
    data=diamonds,
    x="carat",
    y="cut",
    facecolor="white",
    edgecolor="red",
    width=0.25,
    ci=None
)
```

```python
fig,ax=plt.subplots(figsize=(10,10))

sns.barplot(
    data=diamonds,
    x="price",
    y="cut",
    hue="clarity",
    ci=None,
    palette="deep"
)

ax.legend(bbox_to_anchor=(1,1))
```

```python
sns.barplot(
    data=diamonds,
    x="cut",
    y="carat",
    hue="color",
    palette=colors_list
)

plt.legend(bbox_to_anchor=(1,1))
```

# Histograms

```python
sns.histplot(
    data=diamonds,
    x="price"
)
```

```python
sns.histplot(
    data=diamonds,
    x="price",
    bins=20,
    kde=True
)
```

```python
sns.histplot(
    data=diamonds.query("clarity in ['IF','I1']"),
    x="price",
    hue="clarity",
    bins=10,
    kde=True
)
```

```python
sns.histplot(
    data=diamonds.query("clarity in ['IF','I1']"),
    x="price",
    hue="clarity",
    stat="percent",
    bins=10,
    #binwidth=2000,
    kde=True
)
```

```python
sns.histplot(
    data=diamonds,
    x="price",
    y="clarity",
    cbar=True
)
```
