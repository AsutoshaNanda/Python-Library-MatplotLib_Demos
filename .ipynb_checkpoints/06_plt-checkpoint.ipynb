```python
import numpy as np
```

```python
data = [1,5,11,24,36,48,61,67,75,81]

print(data)
```

```python
labels = [f'{x}*' for x in range (10)]

print(labels)
```

```python
data_2nd = [x * 5 for x in range (10)]
print(data_2nd)
```

```python
data_3rd = [x+8 for x in range (10)]
print(data_3rd)
```

```python
fig ,ax = plt.subplots()
ax.plot(labels,data)
ax.plot(labels,data_2nd)
ax.plot(labels, data_3rd)

ax.set_xlabel("labels used" , fontsize =14)
ax.set_ylabel("data overall used " , fontsize = 14)

ax.legend(["data","data2","data3"],loc = 'best',ncol=3,frameon = False)
```
