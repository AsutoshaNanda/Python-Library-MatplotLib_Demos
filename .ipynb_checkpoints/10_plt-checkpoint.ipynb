```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/Users/puchku-home/Study/Python Libs/Data Viz with Matplotlib & Seaborn/Data/housing_data.csv')
```

```python
df.head()
```

```python
df2 = df.groupby("region_name").agg({
    'total_homes_sold': 'sum',
    'inventory': 'sum',
    'active_listings':'sum'
})
```

```python
df2.head()
```

```python
df=df.groupby("region_name").agg({'median_active_list_price': 'sum'})
```

```python
mean_price = df["median_active_list_price"].mean()
mean_price
```

```python
df=df.head()
```

```python
df.head()
```

```python
fig,ax = plt.subplots(figsize=(8,7))
ax.bar(df.index,df["median_active_list_price"]/1000000,color="blue")

plt.xticks(rotation=25)

ax.set_xlabel("Regions---->",fontsize =18, labelpad=20)
ax.set_ylabel("Price in Millions---->",fontsize =18)

fig.suptitle("My First BAR Chart",fontsize =20)
```

```python
fig,ax = plt.subplots(figsize=(8,7))
ax.bar(df.index,df["median_active_list_price"]/1000000)

ax.axhline(mean_price/1000000,color = "red",linewidth =3, linestyle =":",xmin=0.1,xmax=0.8)

plt.xticks(rotation=25)

ax.set_xlabel("Regions---->",fontsize =18, labelpad=20)
ax.set_ylabel("Price in Millions---->",fontsize =18)

fig.suptitle("Horizontal Line Bar Chart",fontsize =20)
```

```python
colors = ["blue","blue","red","green","green"]
fig , ax = plt.subplots(figsize =(8,7))

ax.barh(df.index,df["median_active_list_price"]/1000000,color=colors)

ax.set_ylabel("Regions---->",fontsize =18, labelpad=20)
ax.set_xlabel("Price in Millions---->",fontsize =18,labelpad=20)

fig.suptitle("Horizontal Line Bar Chart",fontsize =20)

ctl1 = mpatches.Patch(color="blue",label="c1")

ctl2= mpatches.Patch(color="red",label = "c2")

ax.legend(handles =[ctl1,ctl2],loc="best")
```

# Stacked Bar Charts

```python
df2 = df2.apply(lambda x : x * 100 / sum(x),axis=1)
```

```python
df2=df2.head()
```

```python
df2.head()
```

```python
fig , ax = plt.subplots()

ax.bar(
    df2.index,
    df2["total_homes_sold"],
    label = 'homes sold',
    color= 'red'
)

ax.bar(
    df2.index,
    df2["inventory"],
    label = 'inventory',
    color= 'blue',
    bottom = df2["total_homes_sold"]
)

ax.bar(
    df2.index,
    df2["active_listings"],
    label = 'active listed',
    color= "green",
    bottom =df2["total_homes_sold"]+df2["inventory"]
)


ax.legend()

ax.set_ylabel("Total houses",fontsize =14)
ax.set_xlabel("Regions",fontsize =14)

plt.xticks(rotation =45)
```

# Grouped Bar Charts

```python
fig , ax =plt.subplots(figsize=(12,7))

width=0.25
x=np.arange(len(df2))

ax.bar(
    x-width,
    df2["total_homes_sold"],
    width=width,
    label ="Total Homes Sold"
    
)

ax.bar(
    x,
    df2["inventory"],
    width = width,
    label = "Inventory",
    color = "green"
)


ax.set_xlabel('Categories',fontsize =14)
ax.set_ylabel('Values',fontsize =14)
ax.set_title('Comparison of Total Homes Sold vs Inventory')

ax.legend()

ax.set_xticks(x)
ax.set_xticklabels(df2.index)
plt.xticks(rotation =45)
```

```python
fig , ax =plt.subplots(figsize=(12,7))

width=0.25
x=np.arange(len(df2))

ax.bar(
    x-width,
    df2["total_homes_sold"],
    width=width,
    label ="Total Homes Sold"
)

ax.bar(
    x,
    df2["inventory"],
    width = width,
    label = "Inventory",
    color = "green"
)

ax.bar(
    x+width,
    df2["active_listings"],
    width = width,
    label = "Acive Listings",
    color = "red"
)

ax.set_xlabel('Categories',fontsize =14)
ax.set_ylabel('Values',fontsize =14)
ax.set_title('Comparison of Total Homes Sold vs Inventory vs Active Listings')

ax.legend()

#ax.set_xticks(x)
#ax.set_xticklabels(df2.index)

plt.xticks(x,df2.index,rotation =45)
```

```python
fig , ax =plt.subplots(figsize=(12,7))

ax.plot(df2.index,df2["total_homes_sold"],color ="blue",label="Homes Sold Total",alpha =1)

ax2=ax.twinx()

width =0.5

ax2.bar(
    df2.index,
    df2["inventory"],
    label ="Inventory Warehouse",
    color="orange",
    width =width,
    alpha  = 0.3
)


ax.set_xlabel('Categories',fontsize =14,labelpad =15)
ax.set_ylabel('Values',fontsize =14)
ax.set_title('Comparison of Total Homes Sold vs Inventory')

fig.legend()
# ax.legend(loc = "upper left")
# ax2.legend()

ax.set_xticks(x)
ax.set_xticklabels(df2.index)
plt.xticks(rotation =45)
```
