```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
data = [1,5,11,24,36,48,61,67,75,81]

print(data)
```

```python
labels = [f'{x}' for x in range (10)]

print(labels)
```

```python
data_2nd = [x * 5 for x in range (10)]
print(data_2nd)
```

```python
data_3rd = [x*x+8 for x in range (10)]
print(data_3rd)
```

```python
data=np.array(data)
data_2nd=np.array(data_2nd)
data_3rd=np.array(data_3rd)
```

```python
fig , ax = plt.subplots()
ax.plot(labels,data/10,label = "data",linestyle = "dotted",linewidth=3,color="red")
ax.plot(labels,data_2nd/10,label="data 2",linestyle = '--',color="black")
ax.plot(labels,data_3rd/10,label = "data3",color = "purple")

fig.suptitle('overall title',fontsize = 18)
ax.set_title('chart title',fontsize=14)

ax.set_xlabel('labels used')
ax.set_title('data involved')

ax.legend(loc="upper left",ncol =3,frameon=True)

ax.set_ylim(0,10)
```

```python
fig , ax = plt.subplots(figsize=(7,6))
ax.plot(labels,data/10,label = "data",linestyle = "dotted",linewidth=3,color="red")
ax.plot(labels,data_2nd/10,label="data 2",linestyle = '--',color="black")
ax.plot(labels,data_3rd/10,label = "data3",color = "purple")

fig.suptitle('overall title applied',fontsize = 16,y=0.97)
ax.set_title('chart title',fontsize=14)

ax.set_xlabel('labels used')
ax.set_title('data involved')

ax.legend(loc="upper left",ncol =3,frameon=True)

ax.set_ylim(0,10)
```
