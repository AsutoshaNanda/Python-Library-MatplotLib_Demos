```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
hotels = pd.read_excel('/Users/puchku-home/Study/Python Libs/Data Viz with Matplotlib & Seaborn/Data/HotelCustomersDataset.xlsx')

hotels.head()
```

```python
fig , ax = plt.subplots(2,3,sharex='all',sharey='all')
ax[0][0].hist(hotels.query("Nationality == 'PRT' and RoomNights < 15").loc[:, "LodgingRevenue"],color="red")
ax[0][0].set_title("PRT")

ax[0][1].hist(hotels.query("Nationality == 'ESP' and RoomNights < 15").loc[:, "LodgingRevenue"])
ax[0][1].set_title("ESP")

ax[0][2].hist(hotels.query("Nationality == 'FRA' and RoomNights < 15").loc[:, "LodgingRevenue"])
ax[0][2].set_title("FRA")

ax[1][0].hist(hotels.query("Nationality == 'DEU' and RoomNights < 15").loc[:, "LodgingRevenue"])
ax[1][0].set_title("DEU")

ax[1][1].hist(hotels.query("Nationality == 'GBR' and RoomNights < 15").loc[:, "LodgingRevenue"])
ax[1][1].set_title("GBR")

fig.suptitle("Top 5 Hotels",fontsize=14,fontweight="bold")

ax[0][0].set_ylabel("Lodging Revenue") 
ax[1][0].set_ylabel("Lodging Revenue") 
plt.tight_layout()

ax[1, 2].set_axis_off()
```
