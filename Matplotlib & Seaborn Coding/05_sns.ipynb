# Reg Plot & LM Plot 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.patches as mpatches
import matplotlib.gridspec as gridspec
```

```python
diamonds = pd.read_csv('/Users/puchku-home/Study/Python Libs/Data Viz with Matplotlib & Seaborn/Data/Diamonds Prices2022.csv')

diamonds.head()
```

```python
sns.regplot(
    data=diamonds,
    x="carat",
    y="price",
)
```

```python
sns.lmplot(
    data=diamonds,
    x="carat",
    y="price",
    hue="clarity"
)
```

```python
sns.lmplot(
    data=diamonds,
    x="carat",
    y="price",
    row="cut",
    col="color"
)

plt.savefig("1st LMplot()")
```

# Joint Plot

```python
sns.jointplot(
    data=diamonds,
    x="carat",
    y="price"
)
```

```python
sns.jointplot(
    data=diamonds,
    x="carat",
    y="price",
    kind="kde"
)
```

```python
sns.jointplot(
    data=diamonds,
    x="carat",
    y="price",
    kind="scatter",
    color="orange",
    #scatter_kws={'alpha':0.04},
    hue="clarity"
)
```

# Pair Plot

```python
sns.pairplot(diamonds.loc[:,["carat","x","y","z"]])
```

```python
sns.pairplot(
    data=diamonds.iloc[:,1:],
    corner=True,
    diag_kind="kde",
    hue="color"
)

plt.savefig("1st Pair Plot")
```
